```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# Generate sample data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.5, random_state=42)

# Create and fit k-Means model
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)

# Predict cluster labels
y_kmeans = kmeans.predict(X)

# Calculate silhouette score for k-Means clustering
score = silhouette_score(X, y_kmeans)
print(f"Silhouette Score for k-Means: {score}")

# Plot clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
plt.title("k-Means Clustering")
plt.show()

# Perform DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Calculate silhouette score for DBSCAN clustering
score_dbscan = silhouette_score(X, y_dbscan)
print(f"Silhouette Score for DBSCAN: {score_dbscan}")

# Plot DBSCAN clusters
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='plasma')
plt.title("DBSCAN Clustering")
plt.show()

```
