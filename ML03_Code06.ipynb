```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
df = pd.DataFrame({'Age': [25, 30, 35, 40, 45],
                   'Salary': [50000, 60000, 65000, 70000, 80000]})

# Scatter plot
plt.scatter(df['Age'], df['Salary'])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()

```

```python
# same with high number of datapoints

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate sample data with a high number of data points
np.random.seed(42)  # For reproducibility
num_data_points = 100
age = np.random.randint(20, 60, num_data_points)
salary = 50000 + (age - 20) * 1000 + np.random.normal(0, 10000, num_data_points)

df = pd.DataFrame({'Age': age, 'Salary': salary})

# Scatter plot
plt.scatter(df['Age'], df['Salary'])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
